**constructing a Box Spread**

1. A **bull call spread**: Buy a call with strike $K_1$ and sell a call with strike $K_2$.

2. A **bear put spread**: Sell a put with strike $K_1$ and buy a put with strike $K_2$.

At maturity, the payoff of this four-leg strategy is always $(K_2 - K_1)$ regardless of the underlying's price, which is why it's often referred to as a (theoretical) risk-free strategy if all options are European. I am using the VIX index (CBOE volatility index), its main advantage is that it is an european option which is crucial for european options, since American options can't lock in the risk-free rate due to early exercise risk and most of the options (ETFs and stocks) on Yahoo Finance are American. If market prices are "out of line", the net cost or net credit is mispriced relative to $(K_2 - K_1)e^{-rT}$, creating an arbitrage.

**The CBOE Volatility Index (^VIX)**

**What it Measures**  
   - A **forward-looking**, model-free gauge of expected **30-day** volatility in the S&P 500 Index (SPX).  
   - Quoted in annualized percentage points: e.g. **20.00** ⇒ ±20 % one-standard-deviation move over the next 30 days.

**How It’s Computed**  
   - Uses a **wide strip** of SPX OTM calls and puts across many strikes to estimate risk-neutral variance.  
   - Key formula for the 30-day variance $\sigma_{30}$:
     $$
       \sigma_{30}^2 
       = \frac{2}{T}\sum_{i}\frac{\Delta K_i}{K_i^2}\,e^{rT}\,Q(K_i)
         \;-\;\frac{1}{T}\Bigl(\frac{F}{K_0}-1\Bigr)^2
     $$
     where  
     - $T = \tfrac{30}{365}$ years  
     - $K_i$ are strikes, $\Delta K_i$ their intervals  
     - $Q(K_i)$ is the mid-quote option price  
     - $F$ is the forward index level, $K_0$ the first strike below $F$  
   - The VIX index value is then $\sqrt{\sigma_{30}^2}\times100$.

**Key Properties**  
   - **Non-directional**: Pools both puts and calls, capturing total market uncertainty.  
   - **Mean-reverting**: Typically trades between 10% – 40% historically.  
   - **Deep liquidity**: Based on highly liquid SPX options—widely used by traders and risk managers.

**Why Trade VIX Options?**  
   - **Hedging**: Protect equity portfolios against volatility spikes.  
   - **Speculation**: Express directional views on volatility itself.  
   - **Diversification**: Volatility often behaves differently from equities.

> **Note:**  
> VIX options are **European-style** and cash-settled on the VIX level at expiration—ideal for multi-leg strategies (e.g. box spreads) without any early-exercise risk.

**c) Price the options with Black Scholes Merton approach to pricing.**

The Black-Scholes-Merton model is used to calculate the theoretical price of European call and put options.

The formulas for the Black-Scholes call and put option prices are:

For a call option:
$$
C = S_0 N(d_1) - K e^{-rT} N(d_2)
$$

For a put option:
$$
P = K e^{-rT} N(-d_2) - S_0 N(-d_1)
$$

where:
$$
d_1 = \frac{\ln(S_0 / K) + (r + 0.5 \sigma^2) T}{\sigma \sqrt{T}}
$$
$$
d_2 = d_1 - \sigma \sqrt{T}
$$

In these formulas:
- ${ S_0 }$ is the current stock price.
- ${ K }$ is the strike price.
- ${ T }$ is the time to maturity.
- ${ r }$ is the risk-free interest rate.
- ${ \sigma }$ is the volatility of the underlying stock.
- ${ N(\cdot) }$ is the cumulative distribution function of the standard normal distribution. `norm.cdf(x)` in python

Implement this logic into Python:


```python
# Black-Scholes utility functions
def black_scholes_call(S, K, T, r, sigma):
    """
    Returns the BSM price of a European call option.
    :param S:     Current underlying price
    :param K:     Strike price
    :param T:     Time to maturity (in years)
    :param r:     Risk-free rate (annualized)
    :param sigma: Volatility (annualized)
    """
    if T <= 0:
        return max(0, S - K)
    d1 = (np.log(S/K) + (r + 0.5*sigma**2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call_price = S*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2)
    return call_price

def black_scholes_put(S, K, T, r, sigma):
    """
    Returns the BSM price of a European put option.
    """
    if T <= 0:
        return max(0, K - S)
    d1 = (np.log(S/K) + (r + 0.5*sigma**2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    put_price = K*np.exp(-r*T)*norm.cdf(-d2) - S*norm.cdf(-d1)
    return put_price
```

**f) Report the payoff at expiration, the profit, and the return.**

Initial net cost (or net credit if negative) is: 

$$Net \:Cost = (long \: call \: at \: K_1) - (short \: call\: at \:K_2) - (short \:put \:at \:K_1) + (long \:put \:at \:K_2)$$

```python
net_cost = long_call_price - short_call_price - short_put_price + long_put_price
print("Net cost of the box spread:", net_cost)
```

**Payoff at Expiration**

The payoff at expiration 𝑇 for a box spread is always $K_2 - K_1$. (Assuming one holds 1 contract of each, ignoring contract multipliers.) So the **intrinsic** payoff is:

$$Payoff(S_T) = K_2 - K_1$$

Hence the profit (ignoring interest rates and time value) is: 

$$Profit = (K_2 - K_1) - (Net Cost)$$

To be fully accurate in real markets, one should discount $(K_2 - K_1)$ by $e^{-rT}$, for the present value. But since the Box Spread should give the implied risk-free rate, I decided not to discount, but to plug the risk-free rate out later on. Each option contract typically covers 100 shares in equity markets, so technically, one has to multiply by 100.

```python
intrinsic_payoff = K2 - K1
profit_at_expiry = intrinsic_payoff - net_cost
print(f"Box spread payoff at expiry = {intrinsic_payoff}")
print(f"Profit (un-discounted) = {profit_at_expiry:.2f}")
```

**Return**

To get the effective yield/ raw return over the time-to-expiry, you’d compare profit to the capital you outlaid: 

$$Return = \frac{(K_2 - K_1) - Net Cost}{Net Cost}$$



In theory, the BSM-based “box cost” should match up with $(K_2 - K_1)e^{-rT}$ if there are no market frictions or arbitrages. 

```python
return_Box_spread = (intrinsic_payoff - net_cost)/net_cost

print(f"raw_return = {return_Box_spread:.2f}%")
```

### Greeks  

| Greek | Call | Put |
|-------|------|-----|
| **Delta**  | $\displaystyle \Delta_C = \Phi(d_1)$ | $\displaystyle \Delta_P = \Phi(d_1)-1$ |
| **Gamma**  | $\displaystyle \Gamma = \frac{\phi(d_1)}{S\,\sigma\sqrt{T}}$ | *same* |
| **Vega**   | $\displaystyle \nu = S\sqrt{T}\,\phi(d_1)$ | *same* |
| **Theta**  | $\displaystyle \Theta_C = -\frac{S\sigma\phi(d_1)}{2\sqrt{T}}\;-\;rK e^{-rT}\Phi(d_2)$ | $\displaystyle \Theta_P = -\frac{S\sigma\phi(d_1)}{2\sqrt{T}}\;+\;rK e^{-rT}\Phi(-d_2)$ |
| **Rho**    | $\displaystyle \rho_C = +K T e^{-rT}\,\Phi(d_2)$ | $\displaystyle \rho_P = -K T e^{-rT}\,\Phi(-d_2)$ |

*All values are **per share**. Multiply by 100 for one listed equity‑option contract.  
$\Theta$ shown above is per **year**; divide by 365 to obtain per‑day decay.*


```python
# Greeks (Delta, Gamma, Vega, Theta, Rho) for calls
def greeks_call(S, K, T, r, sigma):
    """
    Computes the main Greeks for a call option under Black-Scholes.
    Returns dict: {'delta':..., 'gamma':..., 'vega':..., 'theta':..., 'rho':...}
    """
    if T <= 0:
        # At expiration, a "Greek" approach is degenerate.
        # We'll treat them as 0 for simplicity.
        return {'delta': float(S > K), 'gamma': 0.0, 'vega': 0.0, 'theta': 0.0, 'rho': 0.0}

    d1 = (np.log(S/K) + (r + 0.5*sigma**2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)

    delta = norm.cdf(d1)
    gamma = norm.pdf(d1)/(S*sigma*np.sqrt(T))
    vega  = S*norm.pdf(d1)*np.sqrt(T)
    theta = (-S*norm.pdf(d1)*sigma/(2*np.sqrt(T))
             - r*K*np.exp(-r*T)*norm.cdf(d2))
    rho   = K*T*np.exp(-r*T)*norm.cdf(d2)

    return {'delta': delta, 'gamma': gamma, 'vega': vega, 'theta': theta, 'rho': rho}

# Greeks for puts (similarly)
def greeks_put(S, K, T, r, sigma):
    if T <= 0:
        return {'delta': -float(K > S), 'gamma': 0.0, 'vega': 0.0, 'theta': 0.0, 'rho': 0.0}

    d1 = (np.log(S/K) + (r + 0.5*sigma**2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)

    delta = norm.cdf(d1) - 1.0  # or -norm.cdf(-d1)
    gamma = norm.pdf(d1)/(S*sigma*np.sqrt(T))
    vega  = S*norm.pdf(d1)*np.sqrt(T)
    theta = (-S*norm.pdf(d1)*sigma/(2*np.sqrt(T))
             + r*K*np.exp(-r*T)*norm.cdf(-d2))
    rho   = -K*T*np.exp(-r*T)*norm.cdf(-d2)

    return {'delta': delta, 'gamma': gamma, 'vega': vega, 'theta': theta, 'rho': rho}
```

```python
def net_greeks_box_spread(S, K1, K2, T, r, sigma_callK1, sigma_callK2, sigma_putK1, sigma_putK2):
    """
    Sum of Greeks for the 4 legs of the box spread:
     +1 call at K1
     -1 call at K2
     -1 put at K1
     +1 put at K2
    """
    g_callK1 = greeks_call(S, K1, T, r, sigma_callK1)
    g_callK2 = greeks_call(S, K2, T, r, sigma_callK2)
    g_putK1  = greeks_put(S, K1, T, r, sigma_putK1)
    g_putK2  = greeks_put(S, K2, T, r, sigma_putK2)

    # Summation with correct signs
    net = {}
    for greek in ['delta','gamma','vega','theta','rho']:
        net[greek] = ( g_callK1[greek]
                     - g_callK2[greek]
                     - g_putK1[greek]
                     + g_putK2[greek] )
    return net

box_greeks = net_greeks_box_spread(current_price, K1, K2, T, r,
                                   iv_call_K1, iv_call_K2, iv_put_K1, iv_put_K2)
print("Box Spread Greeks (approx.):", box_greeks)
```

### Interpreting the Net Greeks of the Box Spread

| Greek | What it measures | Ideal (pure European box) | The value | Plain‑English meaning |
|-------|------------------|---------------------------|------------|-----------------------|
| **Delta (Δ)** | 1st‑order sensitivity to the underlying price $\frac{∂V}{∂S}$ | 0 | **0.106** | A \$1 stock move changes value by about **10.6 ¢** per share. |
| **Gamma (Γ)** | Curvature (how delta itself changes), ∂²V/∂S² | 0 | **0.009** | Delta shifts by about 0.009 for each \$1 move – tiny. |
| **Vega (ν)**  | Sensitivity to implied vol, ∂V/∂σ | 0 | **-0.19** | A 1 pp (0.01) vol change moves value ≈ **-0.19 ¢** per share. |
| **Theta (Θ)** | Time‑decay (annual), ∂V/∂t | 0 | **0.668** | Per day: 0.668 / 365 ≈ **0.002 ¢** per share – negligible. |
| **Rho (ρ)**   | Sensitivity to the risk‑free rate, ∂V/∂r | 0 | **‑0.057** | A 100 bp rise in rates lowers value by **6 ¢** per share. |

> **Note:** Figures are *per share*. Multiply by 100 for one listed option contract.

---

#### What the Numbers Tell Us

**Price-neutral:** The small delta (≈0.106) shows the box spread isn’t perfectly delta-neutral. Gamma is ≈0, so delta itself barely changes for small moves in the underlying.  
* **Vol‑neutral:** Vega implies sub‑penny changes for realistic volatility shifts.  
* **Minor θ and ρ:** Time decay and rate exposure are only a few cents per contract per day or per 1 % rate shock.

These small residual Greeks arise from

* slight differences in implied volatilities across strikes, and  
* bid/ask asymmetry plus rounding.

Overall, the box behaves like a **synthetic risk‑free bond** paying $K_2 - K_1$ at expiry, with some minor residual exposures.
